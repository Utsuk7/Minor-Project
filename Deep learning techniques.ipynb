```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Assuming your original data is a 1D array with size 10304
# Adjust this assumption based on the actual shape of your data

# Reshape the data into a 3D array
X_train_reshaped = X_train.reshape((1, 5152, 2))
y_train_reshaped = y_train.reshape((1, 5152, 2))

X_val_reshaped = X_val.reshape((1, 1289, 2))
y_val_reshaped = y_val.reshape((1, 1289, 2))

# Define your model
model = Sequential()

# Convolutional layer with a larger kernel_size
model.add(Conv1D(filters=32, kernel_size=5, activation='relu', input_shape=(None, 2)))

# GRU layer without return_sequences
model.add(GRU(units=64))

# Dense layer after GRU
model.add(Dense(16, activation='swish'))

# Dropout layer
model.add(Dropout(0.2))

# Final Dense layer
model.add(Dense(2))

# Print model summary to check the architecture
model.summary()

# Compile the model with Mean Squared Error loss
model.compile(optimizer=Adam(), loss='mean_squared_error')

# Training
epochs = 50
batch_size = 32  # You can adjust this based on your preference

model.fit(X_train_reshaped, y_train_reshaped, epochs=epochs, batch_size=batch_size, validation_data=(X_val_reshaped, y_val_reshaped))
```

```python
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler

# Assuming your model and scaler are named 'model' and 'scaler' respectively, and have been trained

# Initialize an empty list to store predictions
predictions_list = []

# Initialize the input_data array
input_data = np.array([[16.127, 87.139], [21.58, 87.05], [17.31, 86.94], [16.82, 86.94], [20.02, 86.99]])

# Initialize an empty list to store time steps
time_steps = []

# Simulate 100 time steps
for t in range(15):
    # Use the same scaler to scale the input data
    scaled_input_data = scaler.transform(input_data)

    # Reshape the input data to match the model's expected input shape
    input_data_reshaped = input_data.reshape((1, scaled_input_data.shape[0], scaled_input_data.shape[1]))

    # Make the prediction
    prediction = model.predict(input_data_reshaped)

    # Append the prediction to the list
    predictions_list.append(prediction)

    # Update the input_data array by appending the new prediction and removing the first entry
    input_data = np.append(input_data, prediction, axis=0)
    input_data = input_data[1:]

    # Append the current time step to the list
    time_steps.append(t + 1)

# Convert predictions_list to a numpy array for plotting
predictions_array = np.array(predictions_list).squeeze()

# Plotting with Plotly
fig = go.Figure()

for i in range(predictions_array.shape[1]):
    fig.add_trace(go.Scatter(x=time_steps, y=predictions_array[:, i], mode='lines', name=f'Feature {i + 1}'))

# Update x-axis ticks and labels
fig.update_layout(
    xaxis=dict(
        tickvals=list(range(min(time_steps), max(time_steps)+1, 5)),
        ticktext=list(range(min(time_steps), max(time_steps)+1, 5)),
        tickangle=-45
    )
)

fig.update_layout(
    title='Predictions Over Time',
    xaxis_title='Time Step',
    yaxis_title='Predicted Values',
    showlegend=True
)

fig.show()
```
