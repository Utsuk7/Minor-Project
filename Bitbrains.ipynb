```python
scale=MinMaxScaler()
data_scaled=scale.fit_transform(data)
data_scaled=pd.DataFrame(data_scaled,columns=data.columns)
data_scaled
```

```python

xtrain,xtest,ytrain,ytest=train_test_split(data_scaled.drop(columns=["CPU cores","Memory capacity provisioned [KB]","CPU usage [%]","CPU capacity provisioned [MHZ]"]),data_scaled["CPU usage [%]"],test_size=0.2,random_state=2)
model = keras.Sequential([
    # input layer
    keras.layers.Dense(19, input_shape=(7,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    # we use sigmoid for binary output
    # output layer
    keras.layers.Dense(1,)
]
)
```

```python
model.compile(optimizer = 'adam',
             loss = "mean_squared_error",)
# now we fit our model to training data
model.fit(xtrain,ytrain,epochs=10)
```
